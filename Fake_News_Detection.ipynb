re.sub() requires 3 arguments:

    pattern
    replacement string
    input string

\S

Matches any non-whitespace character (letters, numbers, symbols, etc.)

Example: a, 1, . all match, but a space ( ) does not.

\S+

+ means one or more of the preceding element.

So \S+ matches a sequence of non-space characters.

\S+@\S+

Combines two sequences of non-space characters with an @ in the middle.

```python
def wordout(text):
    # convert text to lowercase
    text = text.lower()
    # remove urls (http, https, www)
    text = re.sub(r"https\S+|http\S+|www\S+", '', text)
    # remove emails
    text = re.sub(r"\S+@\S+", '', text)
    # remove HTML tags
    text = re.sub(r"<.*?>", '', text)
    # remove words containing numbers
    text = re.sub(r"\w*\d\w*", '', text)
    # remove punctuation and special characters (keep letters + spaces)
    text = re.sub(r"[^a-z\s]", '', text)
    # remove extra spaces and newlines
    text = re.sub(r'\s+', ' ', text).strip()
    return text
    
```

```python
data['text']=data['text'].apply(wordout)
```

```python
x=data['text']
y=data['class']
```

# train-test split

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

# vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization=TfidfVectorizer()
xv_train=vectorization.fit_transform(x_train)
xv_test=vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(xv_train,y_train)
```

```python
pred_lr=lr.predict(xv_test)
```

```python
lr.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
pred_dt=dt.predict(xv_test)
```

```python
dt.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```
